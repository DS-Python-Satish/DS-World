### DATA VISUALIZATION

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(14,6),dpi=200)

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_ylabel("Sales")
axes[1].set_title("Radio Spend")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_ylabel("Sales")
axes[2].set_title("Newspaper Spend")

plt.tight_layout();
```

```python
# Relationships between features
sns.pairplot(df,diag_kind='kde')
```

#### Train, Test and Split

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
len(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split data 70% for train and 30% for test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(X_train)
```

```python
len(X_test)
```

#### Create a LinearRegression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

#### Fit/Train the Model on the training data

```python
model.fit(X_train,y_train)
```

#### Calculate Performance on Test Set

```python
test_predictions = model.predict(X_test)
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
RMSE = np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
MAE
```

```python
RMSE
```

```python
df['sales'].mean()
```

#### Retraining Model on Full Data

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

#### Coefficients

```python
final_model.coef_
```

```python
coeff_df = pd.DataFrame(final_model.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

#### Prediction on New Data

```python
# Input: total spend of 150k on TV, 30k on Radio, and 10k on Newspaper Ads
campaign = [[150,30,10]]
```

```python
final_model.predict(campaign)
```

#### Model Persistence (Saving and Loading a Model)

```python
from joblib import dump, load
```

```python
dump(final_model, 'sales_model.joblib') 
```

```python
loaded_model = load('sales_model.joblib')
```

```python
loaded_model.predict(campaign)
```
