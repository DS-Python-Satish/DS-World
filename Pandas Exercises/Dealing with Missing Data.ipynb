```python
# Drop column in case of any NA Values
df.dropna(axis=1)
```

```python
#Keep only the rows with at least 4 non-NA values
df.dropna(thresh=4,axis=1)
```

#### Fill Mising Data

```python
df.head()
```

```python
df.fillna("xxxxx")
```

```python
# missing first_names filled with Empty
df['first_name'] = df['first_name'].fillna("Empty")
df.head()
```

```python
#missing post movie scores filled with average value
df['post_movie_score'].fillna(df['post_movie_score'].mean())
```

```python
# Allmissing columns filled with average value
df.fillna(df.mean())
```
