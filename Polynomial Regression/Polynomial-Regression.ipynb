# Polynomial Regression
***In case of data points clearly not fit a linear regression , then its better go for polynomial regression.***

***Polynomial regression is samilar like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [1,3,4,6,7,8,10,13,15,16,17,19,22,23,24,29,30,35]
y=[100,90,80,70,60,65,70,65,60,70,65,76,80,79,90,95,99,100]
```

```python
# Draw a scatter plot
plt.scatter(x, y)
plt.show()
```

```python
# Draw the line of linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

Data points clearly not fit in a linear regression (a straight line through all data points), so ideal for polynomial regression.

```python
#Draw the line of Polynomial Regression
mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(0, 35, 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y, mymodel(x))
```

Result 0.92,so we can use polynomial regression in future predictions.

```python
print(mymodel(18))
```

predicted value to be 70.35, also visible the same in above diagram

####  Plese see the below example for a bad fit from polynomial regression

```python
x = np.arange(18)
np.random.seed(101)
y=np.random.randint(0,100,18)
```

```python
#Draw the line of Polynomial Regression
mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(0, 18, 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y, mymodel(x))
```

Score 0.14 indicates a very bad relationship, and tells us that this data set is not suitable for polynomial regression.
